```python
import tkinter as tk
from tkinter import scrolledtext

# Predefined responses (you can expand this list)
responses = {
    "What is your name?": "My name is ChatBot.",
    "How are you?": "I'm doing well, thank you!",
    "What can you do?": "I can answer basic questions.",
    "Tell me a joke.": "Why don't scientists trust atoms? Because they make up everything!",
    "What's the weather like today?": "I'm sorry, I don't have access to real-time weather information."
}

# Function to get responses
def get_response(question):
    return responses.get(question, "Sorry, I don't understand that question.")

# Function to handle sending messages
def send_message():
    message = entry_box.get()
    if message.strip():
        chat_area.config(state=tk.NORMAL)
        chat_area.insert(tk.END, "You: " + message + "\n")
        answer = get_response(message)
        chat_area.insert(tk.END, "Bot: " + answer + "\n")
        chat_area.config(state=tk.DISABLED)
        chat_area.yview(tk.END)
    entry_box.delete(0, tk.END)

# Create main window
root = tk.Tk()
root.title("Simple Chatbox")
root.configure(bg='#F0EAD6')  # Set background color

# Create chat area
chat_area = scrolledtext.ScrolledText(root, wrap=tk.WORD, state=tk.DISABLED, width=50, height=20, bg='#FFFBF2', fg='#000000')
chat_area.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
chat_area.config(state=tk.NORMAL)
chat_area.insert(tk.END, "Hello! I'm ChatBot. How can I help you today?\n")
chat_area.config(state=tk.DISABLED)

# Create entry box
entry_box = tk.Entry(root, width=40, bg='#FFF8DC')
entry_box.grid(row=1, column=0, padx=10, pady=10)

# Create send button
send_button = tk.Button(root, text="Send", command=send_message, bg='#FFB6C1', fg='#000000')
send_button.grid(row=1, column=1, padx=10, pady=10)

# Predefined questions (optional)
predefined_questions = [
    "What is your name?",
    "How are you?",
    "What can you do?",
    "Tell me a joke.",
    "What's the weather like today?"
]

# Function to handle predefined questions (optional)
def send_predefined_message(question):
    chat_area.config(state=tk.NORMAL)
    chat_area.insert(tk.END, "You: " + question + "\n")
    answer = get_response(question)
    chat_area.insert(tk.END, "Bot: " + answer + "\n")
    chat_area.config(state=tk.DISABLED)
    chat_area.yview(tk.END)

# Create buttons for predefined questions (optional)
questions_frame = tk.Frame(root, bg='#F0EAD6')
questions_frame.grid(row=2, column=0, columnspan=2, pady=10)

for question in predefined_questions:
    question_button = tk.Button(questions_frame, text=question, command=lambda q=question: send_predefined_message(q), bg='#ADD8E6', fg='#000000')
    question_button.pack(side=tk.LEFT, padx=5)

# Run the main loop
root.mainloop()
```
